```python
import numpy as np
```

```python
def split_train_test(data, test_ratio):
    np.random.seed(seed=42)
    shuffle_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    train_indices, test_indices = shuffle_indices[test_set_size:], shuffle_indices[:test_set_size]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, .2)
```

```python
print("train set size: {}, test set size: {}".format(len(train_set), len(test_set)))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

```python
print("(WITH SKLEARN) train set size: {}, test set size: {}".format(len(train_set), len(test_set)))
```

```python
housing.hist(column='median_income',bins=50,figsize=[14,6])
```

```python
housing["income_cat"] = np.ceil(housing["median_income"]/1.5)
```

```python
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
housing["income_cat"].value_counts() / len(housing)
```

```python
for set in (strat_train_set, strat_test_set):
    set.drop(["income_cat"], axis=1, inplace=True)
```

```python
housing = strat_train_set.copy()
```

```python
housing.plot(kind="scatter",x="longitude", y="latitude",alpha=.4,
             s=housing["population"]/100, label="population",
             c=housing["median_house_value"], cmap=plt.get_cmap("jet"), colorbar=True, figsize=[14,6])
```
